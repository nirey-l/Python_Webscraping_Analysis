```python
a = input("문장을 입력하세요: ")

# split()은 앞뒤 공백 제거 + 단어 사이 여러 공백을 한꺼번에 처리해줍니다.
words = a.split() 

# join을 써서 단어 사이 공백을 넣어서 '하나'로 통일된 예쁜 문장이 나옵니다.
# .join(): 여러 개의 문자열 조각들을 특정 문자를 사이에 넣어서 하나의 커다란 문자열로 합칠 때 사용
clean_sentence = " ".join(words)

print(f"공백 제거: {clean_sentence}")
print(f"단어 개수: {len(words)}개")
```

문제 3-3. 학생들의 성적을 딕셔너리로 저장하고 평균 점수를 계산하는 프로그램을 작성하세요.


```python
# 성적 데이터를 딕셔너리로 저장
grades = {
    "김철수": 85,
    "이영희": 92,
    "박민수": 78,
    "최수진": 95
}

print("학생 성적:")

# 반복문을 사용해 학생별 점수 출력
# .items(): 키와 값 모두 사용
for name, score in grades.items():
    print(f"{name}: {score}점")

# 점수들만 뽑아서 합계와 평균 계산
# .values(): 값만 사용
scores = grades.values()
total_sum = sum(scores)
average = total_sum / len(grades)

print(f"평균 점수: {average}점")
```

문제 3-4. 좌표를 튜플로 저장하고 두 점 사이의 거리를 계산하는 프로그램을 작성하세요.


```python
import math

a = (0, 0)
b = (3, 4)

# x좌표끼리의 차이와 y좌표끼리의 차이 구하기
# 튜플에서 값을 꺼낼 때는 인덱스 [0], [1]을 사용
x_diff = b[0] - a[0]
y_diff = b[1] - a[1]

# 거리 계산 
# math.sqrt(): 제곱근
distance = math.sqrt(x_diff**2 + y_diff**2)

print(f'점 1: {a}')
print(f'점 2: {b}')
print(f'두 점 사이의 거리: {distance}')
```

문제 3-5. 쇼핑 카트를 딕셔너리로 구현하여 상품을 추가하고 총 가격을 계산하는 프로그램을 작성하세요.

```python
prices = {"사과": 1000, "바나나": 800, "오렌지": 1500}

# 장바구니에 담은 수량
cart = {"사과": 2, "바나나": 3, "오렌지": 1}

total_price = 0

print("쇼핑 카트:")

# 장바구니 품목을 하나씩 꺼내며 계산 및 출력
for item, count in cart.items():
    # prices 딕셔너리에서 해당 상품의 가격을 가져옴
    price_per_item = prices[item]
    item_total = price_per_item * count
    
    # 총 가격에 더하기
    total_price += item_total
    
    print(f"{item}: {count}개 (개당 {price_per_item}원) = {item_total}원")


print(f"총 가격: {total_price}원")
```

문제 3-6. 리스트에서 최댓값, 최솟값, 두 번째로 큰 값을 찾는 프로그램을 작성하세요.

```python
numbers = [15, 3, 27, 8, 19, 12, 31]

# 최댓값과 최솟값 구하기
max_val = max(numbers)
min_val = min(numbers)

# 두 번째로 큰 값 구하기
# sorted(): 리스트 정렬
# reverse=True: 내림차순(큰 수 -> 작은 수)
sorted_numbers = sorted(numbers, reverse=True)
second_max = sorted_numbers[1]  # 정렬된 리스트에서 두 번째(인덱스 1) 값

print(f"숫자 목록: {numbers}")
print(f"최댓값: {max_val}")
print(f"최솟값: {min_val}")
print(f"두 번째로 큰 값: {second_max}")
```

```python
words = ['cat', 'elephant', 'dog', 'butterfly', 'ant']

# 길이를 기준으로 최댓값과 최솟값을 가진 '단어' 찾기
# key=len을 넣어줘야 글자 수로 비교함
longest_word = max(words, key=len)
shortest_word = min(words, key=len)

# 찾은 단어의 길이를 각각 구하기
max_len = len(longest_word)
min_len = len(shortest_word)

print(f"단어 목록: {words}")
print(f"가장 긴 단어: {longest_word} ({max_len}글자)")
print(f"가장 짧은 단어: {shortest_word} ({min_len}글자)")
```

문제 3-9. 두 개의 리스트를 병합하고 공통 요소를 찾는 프로그램을 작성하세요.

```python
list1 = [1, 2, 3, 4, 5]
list2 = [4, 5, 6, 7, 8]

# 병합된 리스트 (중복 제거 후 정렬)
# 두 리스트를 더한(+) 뒤 set으로 중복을 없애고 정렬
combined_list = sorted(list(set(list1 + list2)))

# 공통 요소
# set의 &(교집합) 연산자를 사용하면 공통된 요소만 찾아줌
common_elements = sorted(list(set(list1) & set(list2)))

print(f"리스트1: {list1}")
print(f"리스트2: {list2}")
print(f"병합된 리스트: {combined_list}")
print(f"공통 요소: {common_elements}")
```

문제 4-5. 사용자가 입력한 숫자가 소수인지 판별하는 프로그램을 작성하세요.

```python
num = int(input("숫자를 입력하세요: "))

# 소수는 1보다 커야 함
if num > 1:
    for i in range(2, num):
        if num % i == 0:
            # 하나라도 나누어떨어지면 소수가 아님
            print(f"{num}은 소수가 아닙니다.")
            break
    # for-else문: for문 안에서 break가 발생하지 않고 반복문이 끝까지 성공적으로 실행되었을 때만 else 블록이 실행
    else:
        # break 문을 만나지 않고 루프가 끝났다면 (즉, 약수가 없다면) 소수
        print(f"{num}은 소수입니다.")
else:
    print(f"{num}은 소수가 아닙니다.")
```

문제 5-1. 문자열을 단어별로 분리했다가 다시 합치는 코드를 split과 join을 사용하여 작성하세요.

```python
text = "Python is awesome programming language"
print(f"원본 문자열: {text}")

words = text.split(" ")
print(f"분리된 단어들: {words}")

# "연결 문구".join()
words = "-".join(words)
print(f"하이픈으로 연결: {words}")

# 대문자로 변환 후 공백으로 연결해서 리스트를 만듦
words = text.upper().split(" ")
# 리스트에 담긴 대문자 단어들 사이에 다시 공백(" ")을 넣어서 하나의 문장으로 합침
words = " ".join(words)
print(f"대문자로 변환 후 공백으로 연결: {words}")
```

```python
fruits = ['사과', '바나나', '오렌지', '포도', '딸기']

print("과일 목록:")

for i, fruit in enumerate(fruits):
    print(f"{i}: {fruit}")
```

문제 5-4. zip을 사용하여 두 리스트를 결합하는 프로그램을 작성하세요.

```python
students = ['김철수', '이영희', '박민수', '최수진']
scores = [85, 92, 78, 95]

print("학생과 점수 매칭:")

for student, score in zip(students,scores):
    print(f"{student}: {score}점")

# 딕셔너리 컴프리헨션으로 딕셔너리 만들기
score_dict = {student: score for student, score in zip(students, scores)}
print(f"점수별 학생 딕셔너리: {score_dict}")
```
