### 1. Daum 뉴스 기사 제목 스크래핑하기

질문1. 아래의 url에서 뉴스 기사의 링크와 제목을 출력하시오
*  경제 뉴스 url = 'https://news.daum.net/economy'

```python
import requests
from bs4 import BeautifulSoup

# dict 타입으로 요청 파라미터 설정
req_param = {
    'sid': 105
}

url = 'https://news.naver.com/section/{sid}'.format(**req_param)
print(url) 

# 요청 헤더 설정 : 브라우저 정보 (사람처럼 보이게 하기 위함)
# 개발자 도구 네트워크 Doc의 헤더에서 가져올 수 있음
req_header = {
    'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/144.0.0.0 Safari/537.36'
}

# requests 의 get() 함수 호출하기 
res = requests.get(url, headers=req_header)

# print(res.text)
# 응답(response)이 OK 이면
# 응답 (response)에서 text 추출
# BeautifulSoup 객체 생성 
print(res.status_code)
print(res.ok)
print(type(res))

#  CSS 선택자
# print(soup.select("div.sa_text a[href*='mnews/article']"))
# *= : 부분 매칭, ^= : 시작 부분 매칭
if res.ok:
    soup = BeautifulSoup(res.text, 'html.parser')
    print(len(soup.select("div.sa_text a[href^='https://n.news.naver.com/mnews/']"))) # <a> 태그 리스트 순회하기 
    # CSS 선택자를 사용해서 a tag 목록 가져오기
    a_tags = soup.select("div.sa_text a[href^='https://n.news.naver.com/mnews/']")
    print(type(a_tags), type(a_tags[0])) # [Tag, Tag]
    for a_tag in a_tags:
        title = a_tag.text.strip() # 여백을 줄일 때는 strip() 시용
        link = a_tag['href'] # 해당하는 링크 가져오기
        print(title, link)
else:
        # 응답(response)이 Error이면 status code 출력
        print(f'Error Code = {res.status_code}')
```
