```python
import pandas as pd

# csv(엑셀)파일을 데이터프레임으로 불러오기
data = pd.read_csv('../data/인구현황.csv')
```

질문1. 지역별 총인구수 Top5

```python
# '전국'이 포함된 0번 행(또는 1000000000 코드 행)을 제외하고 계산
data[data['행정기관'] != '전국'][['행정기관', '총인구수']].sort_values(by='총인구수', ascending=False).head()
```

질문1-1. 지역별 총인구수 plot

```python
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import seaborn as sns

# font name을 알고 있다면 생략가능
# 한글폰트 path 설정
font_path = 'C:\\Windows\\Fonts\\malgunsl.ttf'
# font의 파일정보로 font name 을 알아내기
font_name = fm.FontProperties(fname=font_path).get_name()
# matplotlib의 rc(run command) 함수를 사용해서 font name(Malgun Gothic) 설정
matplotlib.rc('font', family=font_name)

figure,(axes1) = plt.subplots(nrows=1, ncols=1)

# hue: 그래프를 알록달록하게
sns.barplot(data.sort_values(by='총인구수', ascending=False), x='행정기관', y='총인구수', ax=axes1, hue='행정기관')
plt.show()
```
