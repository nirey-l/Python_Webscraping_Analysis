```python
import pandas as pd

# csv(엑셀)파일을 데이터프레임으로 불러오기
# 데이터프레임: 파이썬으로 구현한 엑셀 시트
data = pd.read_csv('../data/인구현황.csv')
```

질문1. 지역별 총인구수 Top5

```python
# '전국'이 포함된 0번 행 또는 1000000000 코드 행을 제외하고 계산
data[data['행정기관'] != '전국'][['행정기관', '총인구수']].sort_values(by='총인구수', ascending=False).head()
```

질문1-1. 지역별 총인구수 plot

```python
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import seaborn as sns

# 한글폰트 path 설정
font_path = 'C:\\Windows\\Fonts\\malgunsl.ttf'
# font의 파일정보로 font name 을 알아내기
font_name = fm.FontProperties(fname=font_path).get_name()
# matplotlib의 rc(run command) 함수를 사용해서 font name(Malgun Gothic) 설정
matplotlib.rc('font', family=font_name)

# '전국' 행 없애기
df_filtered = data.drop(0).sort_values(by='총인구수', ascending=False)

# 그래프 그리기
# plt.subplots(): 그래프가 그려질 '그림(figure)'과 '축(axes1)'을 만듦
figure,(axes1) = plt.subplots(nrows=1, ncols=1)
figure.set_size_inches(15, 5)

# 데이터 정렬 및 시각화
# hue: 그래프를 알록달록하게
# .sort_values(by=열, ascending=True(오름차순),False(내림차순), inplace=True(원본 데이터에 덮어씀),False(원본은 그대로 두고 복사본을 반환))
sns.barplot(data=df_filtered.sort_values(by='총인구수', ascending=False), x='행정기관', y='총인구수', ax=axes1, hue='행정기관')
# x축 이름(label) 제거하기
axes1.set_xlabel('')
plt.title('지역별 총인구수', fontsize=20)
plt.show()
```

```python
# '전국'이 포함된 0번 행 또는 1000000000 코드 행을 제외하고 계산
# .reset_index(): 인덱스 번호 초기화 0번부터 시작
# drop=True: 기존 번호를 삭제하고 새 번호만 남김
data[data['행정기관'] != '전국'][['행정기관', '남여 비율']].sort_values(by='남여 비율', ascending=False).reset_index(drop=True)
```

질문3-1. 남녀 비율 분석 plot
*   기준이 되는 빨간선은 pyplot의 axhline() 함수를 사용해보세요

```python
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import seaborn as sns

# 한글폰트 path 설정
font_path = 'C:\\Windows\\Fonts\\malgunsl.ttf'
# font의 파일정보로 font name 을 알아내기
font_name = fm.FontProperties(fname=font_path).get_name()
# matplotlib의 rc(run command) 함수를 사용해서 font name(Malgun Gothic) 설정
matplotlib.rc('font', family=font_name)

# '전국' 행 없애기
df_filtered = data.drop(0).sort_values(by='남여 비율', ascending=False)

# 그래프 그리기
# plt.subplots(): 그래프가 그려질 '그림(figure)'과 '축(axes1)'을 만듦
figure,(axes1) = plt.subplots(nrows=1, ncols=1)
figure.set_size_inches(15, 5)

# 데이터 정렬 및 시각화
# hue: 그래프를 알록달록하게
# .sort_values(by=열, ascending=True(오름차순),False(내림차순), inplace=True(원본 데이터에 덮어씀),False(원본은 그대로 두고 복사본을 반환))
sns.barplot(data=df_filtered.sort_values(by='남여 비율', ascending=False), x='행정기관', y='남여 비율', ax=axes1, hue='행정기관')
# 기준이 되는 빨간선
# y=1.0 위치에 빨간색(red), 파선(linestyle='--')의 선
axes1.axhline(y=1.0, color='r', linestyle='--', linewidth=2)
# x축 이름(label) 제거하기
axes1.set_xlabel('')
plt.title('지역별 남녀 비율', fontsize=20)
plt.show()
```

질문4. 남자초과/여자초과 분석

```python
# '남초여초' 판별 조건 추가
# '남여 비율' 컬럼을 기준으로 새로운 값을 할당
data.loc[data['남여 비율'] > 1, '남초여초'] = '남초'
data.loc[data['남여 비율'] < 1, '남초여초'] = '여초'
data.loc[data['남여 비율'] == 1, '남초여초'] = '동일'

# '전국'이 포함된 0번 행 또는 1000000000 코드 행을 제외하고 계산
# .reset_index(): 인덱스 번호 초기화 0번부터 시작
# drop=True: 기존 번호를 삭제하고 새 번호만 남김
result = data[data['행정기관'] != '전국'][['행정기관', '남여 비율', '남초여초']].reset_index(drop=True)
# 인덱스 번호가 1부터 나오게 하기
result.index = result.index + 1
result
```

질문5. 세대당 인구 평균보다 높은 지역

```python
# 인덱스 번호 대신 '행정기관' 컬럼 값이 '전국'인 행을 찾아서 값을 가져오기
# .values[0]: 데이터프레임 형태가 아닌 '값' 자체만 빼오기
total_avg = data[data['행정기관'] == '전국']['세대당 인구'].values[0]

# '전국'이 포함된 0번 행 또는 1000000000 코드 행을 제외하고 계산
# .reset_index(): 인덱스 번호 초기화 0번부터 시작
# drop=True: 기존 번호를 삭제하고 새 번호만 남김
result = data[data['행정기관'] != '전국'][['행정기관', '세대당 인구']].sort_values(by='세대당 인구', ascending=False).reset_index(drop=True).head(7)

print(f"전국 평균 세대당 인구수: {total_avg}")
display(result)
```

질문5-1. 남초 vs 여초 개수 Plot (seaborn의 countplot)  
지역별 세대수 Plot (seaborn의 barplot)

```python
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import seaborn as sns

# 한글폰트 path 설정
font_path = 'C:\\Windows\\Fonts\\malgunsl.ttf'
# font의 파일정보로 font name 을 알아내기
font_name = fm.FontProperties(fname=font_path).get_name()
# matplotlib의 rc(run command) 함수를 사용해서 font name(Malgun Gothic) 설정
matplotlib.rc('font', family=font_name)

# '전국' 행 제외
df_filtered = data[data['행정기관'] != '전국'].copy()

# '남초여초' 판별 조건 추가
# '남여 비율' 컬럼을 기준으로 새로운 값을 할당
df_filtered.loc[df_filtered['남여 비율'] > 1, '남초여초'] = '남초'
df_filtered.loc[df_filtered['남여 비율'] < 1, '남초여초'] = '여초'
df_filtered.loc[df_filtered['남여 비율'] == 1, '남초여초'] = '동일'

# 그래프 그리기
# plt.subplots(): 그래프가 그려질 '그림(figure)'과 '축(axes1)'을 만듦
figure, (axes1,axes2) = plt.subplots(nrows=2, ncols=1)
figure.set_size_inches(15, 5)

# 데이터 정렬 및 시각화
# hue: 그래프를 알록달록하게
# .sort_values(by=열, ascending=True(오름차순),False(내림차순), inplace=True(원본 데이터에 덮어씀),False(원본은 그대로 두고 복사본을 반환))
sns.countplot(data=df_filtered, x='남초여초', ax=axes1, order=['여초', '동일', '남초'], hue='남초여초')
axes1.set_title('남초 vs 여초 지역 개수', fontsize=10)
sns.barplot(data=df_filtered.sort_values(by='세대수', ascending=False), x='행정기관', y='세대수', ax=axes2, hue='행정기관')
axes2.set_title('지역별 세대수', fontsize=10)
# 그래프 간 간격 자동 조절
plt.tight_layout()
plt.show()
```

질문6. 남자 인구 vs 여자 인구 비교 Plot

```python
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import seaborn as sns

# 한글폰트 path 설정
font_path = 'C:\\Windows\\Fonts\\malgunsl.ttf'
# font의 파일정보로 font name 을 알아내기
font_name = fm.FontProperties(fname=font_path).get_name()
# matplotlib의 rc(run command) 함수를 사용해서 font name(Malgun Gothic) 설정
matplotlib.rc('font', family=font_name)

# '전국' 행을 제외하고 분석에 필요한 '행정기관', '남자 인구수', '여자 인구수' 3개의 열만 선택
# .copy(): 원본을 건드리지 않고 복사본을 만들어 안전하게 작업
df_filtered = data[data['행정기관'] != '전국'][['행정기관', '남자 인구수', '여자 인구수']].copy()

# seaborn의 hue 기능을 쓰기 위해 '남자 인구수'와 '여자 인구수'를 하나의 열로 합침
df_melted = df_filtered.melt(id_vars='행정기관', var_name='성별', value_name='인구수')

# 그래프 그리기
# plt.subplots(): 그래프가 그려질 '그림(figure)'과 '축(axes1)'을 만듦
figure, axes1 = plt.subplots(nrows=1, ncols=1)
figure.set_size_inches(15, 5)

sns.barplot(data=df_melted, x='행정기관', y='인구수', hue='성별', ax=axes1,
            palette={'남자 인구수': 'blue', '여자 인구수': 'red'})
plt.title('남자 vs 여자 인구 비교', fontsize=10)
# x축의 행정기관 이름을 45도 회전
plt.xticks(rotation=45)
plt.show()
```
